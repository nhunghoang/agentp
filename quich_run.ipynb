### Import agentp and create the Project

```python
from agentp import Project, SummaryTWAS, IndividualTWAS, GWAS
import os

repo_root = "/Users/zoeytang/Desktop/agentp/example"
main_project_folder = os.path.join(repo_root, "example_project")

project = Project(main_project_folder)
```

### Summary TWAS (JTI hippocampus) + add GWAS

```python
stwas = SummaryTWAS(project, "JTI", "hippocampus")

stwas.add_gwas(
    "amygdala_volume",
    "example/example_data/vol_mean_amygdala.regenie"
)
```

### Run summary TWAS

```python
stwas.run_twas("amygdala_volume")
```

### Individual TWAS

```python
gtwas = IndividualTWAS(project, "FUS", "caudate")
```

### Add subject/covariate/phenotype/genotype files

```python
project.set_subjects("example/example_data/subjects.txt")
project.add_covariates("example/example_data/covariates.csv")
project.add_phenotypes("example/example_data/volumes.csv")

project.add_genotypes("example/example_data/genotypes", "test_c*.bgen")
```

### Predict GReX

```python
gtwas.predict_grex()
```

### Run individual TWAS

```python
gtwas.run_twas("putamen_volume")
```

### Run GWAS on hippocampus_volume

```python
gwas = GWAS(project)
gwas.run("hippocampus_volume")
```
